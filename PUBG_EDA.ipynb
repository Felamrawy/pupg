## Dataset Overview

The dataset comprises 44 observations and 20 variables, representing weapon characteristics.

* **Critical Issues:**
    * **Missing Values:** Bullet Type, Range, and Damage profiles.
    * **Redundancy:** Derived variables may overlap.
    * **Encoding:** *Bullet Type* is incorrectly numerical. *Fire mode* has multiple enteries 

*Note*: Technical definitions, units, and calculations are currently unavailable, which may impact the precision of result interpretations.

**Handling Missing data strategy**:

Most of missing values are structurally dependent on weapon type (e.g., non-firearm weapons lack ammunition-related attributes). Other missing values were cross-validated using external domain sources. This preserves dataset completeness while avoiding bias introduced by listwise deletion, which would disproportionately remove specific weapon categories.

   




------------------------------

### Data Cleaning  

```python
print("Unique values in Bullet types column:")
df['Bullet Type'].unique()
```

```python
# Bullet Type to categorical
df['Bullet Type'] = df['Bullet Type'].astype('object')
bulletmap = {
    5.56: '5.56mm',
    7.62: '7.62mm',
    9.0: '9mm',
    0.45: '.45 ACP',
    12: '12 Gauge',
    0.3: '.300 Magnum'
}
# Replace numeric ammo with labels
df['Bullet Type'] = df['Bullet Type'].replace(bulletmap)

# Fill missing bullet types
df['Bullet Type'] = df['Bullet Type'].fillna('No Ammo').astype('category')
```

### Handling NaN

Instead of removing observations with missing values, a domain-informed imputation strategy was adopted:

- Weapons were categorised into firearm vs non-firearm


```python
def classify_weapon(row):
    if row['Bullet Type'] == 'No Ammo':
        return 'Non-Firearm'
    else:
        return 'Firearm'

df['Weapon Class'] = df.apply(classify_weapon, axis=1)
```

- Missing numerical values were imputed using verified external references

- Sawed-Off, https://pubg.fandom.com/wiki/Sawed-off (**Bullet speed : 330**)    
- R45, https://pubg.fandom.com/wiki/R45 (**Bullet speed : 330**)

- Skorpion, https://pubg.fandom.com/wiki/Skorpion
  
| Level | BDMG (Body) | HDMG (Head) |
| :--- | :--- | :--- |
| Level 0 | 22.0 | 44.0 |
| Level 1 | 15.40 | 30.80 |
| Level 2 | 13.20 | 26.40 |
| Level 3 | 9.90 | 19.80 |


```python

# Bullet speed (m/s)
df.loc[df['Weapon Name'] == 'Sawed-Off', 'Bullet Speed'] = 330
df.loc[df['Weapon Name'] == 'R45', 'Bullet Speed'] = 330

# Skorpion body damage
df.loc[df['Weapon Name'] == 'Skorpion', ['BDMG_0','BDMG_1','BDMG_2','BDMG_3']] = [22.0, 15.40, 13.20, 9.90]

# Skorpion head damage
df.loc[df['Weapon Name'] == 'Skorpion', ['HDMG_0','HDMG_1','HDMG_2','HDMG_3']] = [44.0, 30.80, 26.40, 19.80]
```

### Fire Mode handling

```python
fire_modes = df['Fire Mode'].str.split(', ')

from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()
fire_mode_dummies = pd.DataFrame(
    mlb.fit_transform(fire_modes),
    columns=mlb.classes_,
    index=df.index
)

# Merge back
df = pd.concat([df, fire_mode_dummies], axis=1)

df['num_fire_modes'] = fire_modes.apply(len)
```

Adding features sutainability
 


```python
df['Damage_per_mag'] = df['Damage'] * df['Magazine Capacity']
df['Kills_per_mag'] = df['Magazine Capacity'] / df['Shots to Kill (Chest)']
```

Time to Kill (TTK) ?

```python
df['TTK_chest'] = (df['Shots to Kill (Chest)'] - 1) * df['Rate of Fire']
    
```

```python
fig = make_subplots(
    rows=2, cols=2, 
    subplot_titles=("No Vest (Level 0)", "Level 1 Vest", "Level 2 Vest", "Level 3 Vest"),
    shared_yaxes=True 
)

#  i to row and col mapping: 0->(1,1), 1->(1,2), 2->(2,1), 3->(2,2)
for i in range(4):
    row = (i // 2) + 1
    col = (i % 2) + 1
    
    fig.add_trace(
        go.Bar(
            x=ammu['Weapon Name'],
            y=ammu[f'BDMG_{i}'].sort_values(),
            name=f'Level {i}',
            showlegend=False 
        ),
        row=row, col=col
    )

fig.update_layout(
    height=800, 
    width=1000, 
    title_text="Body Damage Comparison by Armor Level",
    template="plotly_white"
)

fig.update_xaxes(tickangle=45)

fig.show()
```
